# Koordináta-rendszerek bázisai

- a Descartes-féle koordinátarendszerben az $\hat{i}$ és a $\hat{j}$ a két bázisvektorunk, ezek alkotják a bázist

```
axes = plt.axes()
axes.set_aspect(aspect=1)
axes.arrow(0,0,1,0, length_includes_head=True, head_width=0.2, head_length=0.4)   
plt.text(0.5,-0.5,'i hat')
axes.arrow(0,0,0,1, length_includes_head=True, head_width=0.2, head_length=0.4)
plt.text(-0.75,0.5,'j hat')

plt.grid()                      
plt.xlim(-2,2)                   
plt.ylim(-2,2)
plt.show()
```

Tulajdonképpen minden vektor előállítható ezen bázisvektorok skalárszorosainak összege által. Tehát például: 
$$a=\begin{bmatrix}
1 \\
2
\end{bmatrix}=1\cdot \hat{i} + 2\cdot \hat{j}$$

Új bázis használata:

```
axes = plt.axes()                                                                    
axes.set_aspect(aspect=1)

#két új bázisvektor
axes.arrow(0,0,2,4, length_includes_head=True, head_width=0.2, head_length=0.4)   
axes.arrow(0,0,4,-2, length_includes_head=True, head_width=0.2, head_length=0.4)

#egy vektor
axes.arrow(0,0,3,1, length_includes_head=True, head_width=0.2, head_length=0.4)
plt.text(3.5,1,'c')

plt.grid()                      
plt.xlim(-6,6)                   
plt.ylim(-6,6)
plt.show()
```

Ha a *c* vektort akarjuk akarjuk meghatározni a másik kér bázisvektort által, akkor annak koordinátái nem $\begin{bmatrix}
3 \\
1
\end{bmatrix}$ lesz, hiszen ezek az eredeti $\hat{i}$ és $\hat{j}$ bázisvektorokkal meghatározott bázisban ennyi. Az új bázisunkban ez: $\begin{bmatrix}
0,5 \\
0,5
\end{bmatrix}$, hiszen a következő, vektorok által felírt egyenletet megoldva (ami nem más mint két vektor lineáris kombinációja):
$$\begin{bmatrix}
3 \\
1
\end{bmatrix}=c_1\begin{bmatrix}
2 \\
4
\end{bmatrix}+c_2 \begin{bmatrix}
4 \\
-2
\end{bmatrix}$$

A $c_1=0,5$ és a $c_2=0,5$ eredményeket kapjuk, mindkét bázisvektorunkon 0,5 egységnyit kell lépnünk, hogy megkapjuk a $c$ vektorunkat. Tehát a tanulság az, hogy egy vektor koordinátái attól függenek, hogy milyen bázisban írjuk le őket. 


Kollineáris vektorok esete:

```
axes = plt.axes()                                                                    
axes.set_aspect(aspect=1)

#két új bázisvektor
axes.arrow(0,0,2,4, length_includes_head=True, head_width=0.2, head_length=0.4)   
axes.arrow(0,0,1,2, length_includes_head=True, head_width=0.2, head_length=0.4)

#egy vektor
axes.arrow(0,0,3,1, length_includes_head=True, head_width=0.2, head_length=0.4)
plt.text(3.5,1,'c')

plt.grid()                      
plt.xlim(-6,6)                   
plt.ylim(-6,6)
plt.show()
```

Néhány G2-es elméleti órai állítás:



*   végesen generált vektortérben bármely két bázis azonos tagszámú $\rightarrow$ igaz, hiszen ugyanannyi tagszámú lehet csak lineárisan független
*   végesen generált vektortér dimenzióján bázisainak közös tagszámát értjük $\rightarrow$ igaz, 2D-ben kéttagú a bázis, 3D-ben háromtagú



# **Lineáris transzformációk**





Az órán elhangzott definíció így szól:

Legyenek $V_1$ és $V_2$ ugyanazon *T* test feletti vektorterek. Ekkor a $\varphi:V_1\rightarrow V_2$ leképezést lineáris leképezésnek nevezzük, ha $\forall \mathbf{a},\mathbf{b}\in V_1$ és $\forall \alpha \in T$ esetén:

*   $\varphi(\mathbf{a}+\mathbf{b})=\varphi(\mathbf{a})+\varphi(\mathbf{b})$
*   $\varphi(\alpha \cdot \mathbf{a})=\alpha \cdot \varphi(\mathbf{a})$

Akkor nevezzük ezt a leképezést lineáris transzformációnak, ha $V_1=V_2$, tehát ugyanazon vektortérben maradunk.

Két új bázisvektorunk van $\begin{bmatrix}
4\\
-2
\end{bmatrix}$ és $\begin{bmatrix}
2 \\
4
\end{bmatrix}$. Most ebben a bázisban írjuk le a vektorokat. Már kiszámoltuk, hogy a $c$ vektorunk koordinátái ebben az új koordinátarendszerben: $\begin{bmatrix}
0,5 \\
0,5
\end{bmatrix}$. Tegyük most egymás mellé a két bázisvektorunkat:
$\begin{bmatrix}
4&2\\
-2 &4
\end{bmatrix}$

Ez a 2x2-es mátrix lesz ennek a transzformációnak a mátrixa.

Írjuk le általánosan ezt transzformációt:
$$\begin{bmatrix}
x\\
y
\end{bmatrix}\rightarrow x \begin{bmatrix}
4\\
-2
\end{bmatrix}+y\begin{bmatrix}
2\\
4
\end{bmatrix}=\begin{bmatrix}
4\cdot x+2\cdot y\\
-2\cdot x+4\cdot y
\end{bmatrix}=\begin{bmatrix}
4&2\\
-2 &4
\end{bmatrix} \begin{bmatrix}
x\\
y
\end{bmatrix}$$



Példa:

Vektor: $\begin{bmatrix}
1 \\
3
\end{bmatrix}$

Transzformált vektor koordinátái: $\begin{bmatrix}
4&2\\
-2 &4
\end{bmatrix}\begin{bmatrix}
1\\
3
\end{bmatrix}=\begin{bmatrix}
10 \\
10
\end{bmatrix}$ 

```
axes = plt.axes()                                                                     
axes.set_aspect(aspect=1)

#két új bázisvektor
axes.arrow(0,0,2,4, length_includes_head=True, head_width=0.2, head_length=0.4)   
axes.arrow(0,0,4,-2, length_includes_head=True, head_width=0.2, head_length=0.4)

#eredeti vektor
axes.arrow(0,0,1,3, length_includes_head=True, head_width=0.2, head_length=0.4)
plt.text(2,5,'a')

#transzformált vektor
axes.arrow(0,0,10,10, length_includes_head=True, head_width=0.2, head_length=0.4)
plt.text(5,4,'transformed')

plt.grid()                      
plt.xlim(-2,11)                   
plt.ylim(-2,12)
plt.tight_layout()
plt.show()
```

Most vizsgáljuk meg, hogy lineáris-e ez a transzformáció!

$\varphi(\mathbf{a}+\mathbf{b})=\begin{bmatrix}
4&2\\
-2 &4
\end{bmatrix}\left(\begin{bmatrix}
a_x\\
a_y
\end{bmatrix}+\begin{bmatrix}
b_x\\
b_y
\end{bmatrix}\right)=\begin{bmatrix}
4a_x+4b_x+2a_y+2b_y\\
-2a_x+-2b_x+4a_y+4b_y
\end{bmatrix}$

$\varphi(\mathbf{a})+\varphi(\mathbf{b})=\begin{bmatrix}
4&2\\
-2 &4
\end{bmatrix}\begin{bmatrix}
a_x\\
a_y
\end{bmatrix}+\begin{bmatrix}
4&2\\
-2 &4
\end{bmatrix}\begin{bmatrix}
b_x\\
b_y
\end{bmatrix}=\begin{bmatrix}
4a_x+2a_y+4b_x+2b_y\\
-2a_x+4a_y-2b_x+4b_y
\end{bmatrix}$

$\varphi(\alpha \cdot \mathbf{a})=\begin{bmatrix}
4&2\\
-2 &4
\end{bmatrix}\left(\alpha\cdot \begin{bmatrix}
a_x\\
a_y
\end{bmatrix}\right)=\begin{bmatrix}
4\alpha a_x +2 \alpha a_y\\
-2\alpha a_x +4\alpha a_y
\end{bmatrix}$

$\alpha \cdot \varphi(\mathbf{a})=\alpha \cdot\begin{bmatrix}
4&2\\
-2 &4
\end{bmatrix}\begin{bmatrix}
\alpha 4 a_x +\alpha 2 a_y\\
\alpha (-2) a_x +\alpha 4 a_y
\end{bmatrix}$

Láthatjuk, hogy mindkettő feltétel teljesül, tehát ez a transzformáció lineáris.

Ellenőrizzük le számszerűen két vektorral!

$\varphi(\mathbf{a}+\mathbf{b})=\begin{bmatrix}
4&2\\
-2 &4
\end{bmatrix}\left(\begin{bmatrix}
1\\
3
\end{bmatrix}+\begin{bmatrix}
2\\
3
\end{bmatrix}\right)=\begin{bmatrix}
24\\
18
\end{bmatrix}$

$\varphi(\mathbf{a})+\varphi(\mathbf{b})=\begin{bmatrix}
4&2\\
-2 &4
\end{bmatrix}\begin{bmatrix}
1\\
3
\end{bmatrix}+\begin{bmatrix}
4&2\\
-2 &4
\end{bmatrix}\begin{bmatrix}
2\\
3
\end{bmatrix}=\begin{bmatrix}
10\\
10
\end{bmatrix}+\begin{bmatrix}
14\\
8
\end{bmatrix}=\begin{bmatrix}
24\\
18
\end{bmatrix}$

$\varphi(\alpha \cdot \mathbf{a})=\begin{bmatrix}
4&2\\
-2 &4
\end{bmatrix}\left(3\cdot \begin{bmatrix}
1\\
3 
\end{bmatrix}\right)=\begin{bmatrix}
30\\
30
\end{bmatrix}$

$\alpha \cdot \varphi(\mathbf{a})=3\cdot\begin{bmatrix}
4&2\\
-2 &4
\end{bmatrix}\begin{bmatrix}
1\\
3 
\end{bmatrix}=\begin{bmatrix}
30\\
30
\end{bmatrix} $
