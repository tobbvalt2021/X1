# Python kód használata google colabban

```
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```
X=[]
Y=[]

for j in range(3000):
  X.append(j*np.cos(j))
  Y.append(j*np.sin(j))

plt.rcParams["figure.figsize"] = [10,10]
plt.scatter(X,Y,s=10.25)
plt.show()
```

```
X=[]
Y=[]

for j in range(1,10000):
  pr=True
  if j>1:
    for k in range(2,int(np.sqrt(j))+1):
      if j%k==0:
        pr=False
    if pr==True:
      X.append(j*np.cos(j))
      Y.append(j*np.sin(j))
  else:
    X.append(j*np.cos(j))
    Y.append(j*np.sin(j))

plt.rcParams["figure.figsize"] = [10,10]
plt.scatter(X,Y,s=5.25)
plt.show()
```
