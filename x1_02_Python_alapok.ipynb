# **Fő adattípusok, amelyeket használni fogunk:**

Numerikus adatokat esetén a lehetőségek:


*   integer - egész számok (pl.: 1, 20, 300)
*   float - lebegőpontos számok (pl.: 3.14, 2.22e-3)
*   complex (pl.: 1j, 3+2j)

```
a=1
print(type(a))  #integer
b=3.14
print(type(b))  #float
c=1j
print(type(c))  #complex
```

```
print(5+3)
print(10-6)
print(4/2)  #2.0
print(5%2)  #1
print(2**4)
print(17//10)
```

Szekvenciális tárolás:



*   list - lista (pl.: grades = [1, 2, 3, 4, 5])
*   tuple - lényegében egy megváltoztathatatlan lista (pl.: grades = (1, 2, 3, 4, 5) )

```
grades=[1,2,3,4,5]
print(grades[4])
print(grades[0:2])
print(len(grades))
grades.append(6)
print(grades[:])
del grades[5]
print(grades)

grades2=(5,4,3,2,1)
#grades2.append(6)    
print(grades2[:4])
```

# **Feladat**


```
def num_digit(x):
  #153//10->15,3->15
  #15//10->1,5->1
  #1//10->0,1->0
  n=0
  while(x!=0):
    x//=10
    n+=1
  return n

def Armstrong_num(x):
  n=num_digit(x)
  sum=0
  #153%10->3
  #153//10->15
  #15%10->5
  #15//10->1
  #1%10->1
  #1//10->0
  temp=x
  while(temp!=0):
    remainder=temp%10
    sum=sum+remainder**n
    temp//=10
  return(sum==x)

x=153
print(Armstrong_num(x))
```
