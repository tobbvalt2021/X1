```
#plotolás objektumok létrehozásával
x1=np.linspace(-50,0,250)
x2=np.linspace(0.1,50,249)
y1=np.cos(x1)
y2=1+1/3*np.sin(x2)
plt.rcParams["figure.figsize"]=[20,8] #plotok átméretezése
fig,ax=plt.subplots() #fig -> rajzfelület; ax -> ábra
ax.scatter(x1,y1,s=20,c='b',marker='s',label='y1') #c -> szín; marker -> jelölőtípus (s: négyzet; o: kör)
ax.scatter(x2,y2,s=10,c='r',marker='o',label='y2')
ax.legend(loc='upper left') #jelmagyarázat áthelyezése
plt.show()
```

```
x1=np.arange(0.0,5.0,0.01)
y1=1+np.sin(x1*np.pi)
x2=np.linspace(0,5,25)
y2=1+np.sin(x2*np.pi)
fig,(ax1,ax2)=plt.subplots(1,2) #egy rajzfelület, két ábra; (1,2) -> 1 sorban és 2 oszlopban
ax1.plot(x1,y1)
ax1.set(xlabel='time[s]',ylabel='force[N]',title='Plot1') #tengelycímek, cím
ax2.plot(x2,y2)
ax2.set(title='Plot2')
ax2.grid()
plt.show()
```

# A tudományos scipy

```
import scipy.linalg as la #egy almodul a sokból
```

```
A=np.array([[1,2],[2,1]])
print(A)
```

```
evalues=la.eig(A) #sajátértékek
print(evalues[0]) #0-s index, ami a többin van, az most nem izgalmas
```

```
print(la.det(A)) #determináns
```

```
print(la.inv(A)) #inverz mátrix
```

```
print(A@la.inv(A)) #szorzatuk egységmátrix -> valóban helyes az inverz
```

# A bevezető óra kódjai

Egyfajta 2D számábrázolás, ahol j -> (xj,yj), egyfajta polárkoordinátás felírás (ld. kód). A spirál alak egyszerűen következik a felírásból (bővebb magyarázat a videóban). A második kód ugyanígy ábrázolja a megadott intervallum prímjeit.

```
import numpy as np
import matplotlib.pyplot as plt
X=[]
Y=[]
for j in range(60):
  X.append(j*np.cos(j))
  Y.append(j*np.sin(j))
plt.rcParams["figure.figsize"]=[10,10]
plt.scatter(X,Y,s=10.25)
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
X=[]
Y=[]
for j in range(1,10000):
  pr=True
  if j>1:
    for k in range(2,int(np.sqrt(j))+1):
      if j%k==0:
        pr=False
    if pr==True:
      X.append(j*np.cos(j))
      Y.append(j*np.sin(j))
plt.rcParams["figure.figsize"]=[10,10]
plt.scatter(X,Y,s=10.25)
plt.show()
```
